```python
#define image1 here
photo_matrix = [[141, 0 , 138, 143, 143, 143, 144],
                [141, 140, 140, 142, 142, 143, 143],
                [140, 145, 145, 144, 142, 142, 145],
                [141, 141, 141, 143, 142, 141, 143],
                [139, 140, 141, 141, 142, 143, 144]]

# Create a single-channel (grayscale) image from the matrix
image1 = np.array(photo_matrix, dtype=np.uint8)

# save the image1
cv2.imwrite('image1.png', image1)
```

Histogram Function

```python
#code here
#first define a function for calculating histogram
#you are free to use libraries
def calc_hist(image):
    '''
    you are free to use libraries
    calculate image histogram
    input(s):
      image (ndarray): input image
    output(s):
      hist (ndarray): computed input image histogram
    '''
    # image, channel index, specific regions, number of bins, range of pixel values
    hist = cv2.calcHist([image], [0], None, [256], [0, 256])

    # Convert histogram to numpy array
    hist_array = np.squeeze(hist)
    print('hist_array: ', np.array(hist_array))
    
    return(hist_array)
```

```python
#dont change this cell
print('image1: ', image1)
plt.imshow(image1, cmap='gray', vmin=0, vmax=255)
plt.figure()
image1_histogram = calc_hist(image1)
plt.plot(image1_histogram)
plt.title('histogram image1')
```

Histogram Clip Function

```python
#code here
#define a function (Clip) for Clipping(input:image , output: Clipped image)

def clip_hist(image, min_value, max_value):
  '''
  don't use libraries
  input(s):
    image (ndarray): input image
    min_value : min value of the histogram which you wanna clip.
    max_value : max value of the histogram which you wanna clip.
  output(s):
    output_image (ndarray): enhanced image with histogram clipping
  '''
  output_image = image.copy()
  # Start
  
  for row in output_image:
    for i in range(len(row)):
      if(row[i]<min_value):
        row[i] = min_value
      elif(row[i]>max_value):
        row[i] = max_value

  # End
  return output_image
```

```python
min_value = 138 # your min value here
max_value = 145 # your max value here

clipped_image = clip_hist(image1, min_value, max_value)
plt.imshow(clipped_image, cmap='gray', vmin=0, vmax=255)
plt.figure()
plt.plot(calc_hist(clipped_image))
plt.title('clipped histogram')

print(clipped_image)
#dont change this cell
# min_value = # your min value here
# max_value = # your max value here
# plt.imshow(clip_hist(image1),cmap='gray',vmin=0,vmax=255)
# plt.figure()
# plt.plot(calc_hist(clip_hist(image1,min_value, max_value)))
# plt.title('clipped histogram')
```

Histogram Stretch Fuction

```python
#code here
#define a function (stretch) for stretching(input:image , output: stretched image)

def stretch_hist(image):
  '''
  don't use libraries
  input(s):
    image (ndarray): input image
  output(s):
    output_image (ndarray): enhanced image with histogram stretching
  '''
  output_image = image.copy()
  # Start
  
  Min_Pixel_Value = np.min(image)
  Max_Pixel_Value = np.max(image)
  New_Min = 0
  New_Max = 255

  for row in output_image:
    for i in range(len(row)):
      x = ((row[i]-Min_Pixel_Value)/(Max_Pixel_Value-Min_Pixel_Value)) * (New_Max-New_Min) + New_Min
      row[i] = x

  # End
  return output_image
```

```python
stretched_image = stretch_hist(clipped_image)
plt.imshow(stretched_image, cmap='gray', vmin=0, vmax=255)
plt.figure()
plt.plot(calc_hist(stretched_image))
plt.title('stretched histogram')

print(stretched_image)
#dont change this cell
# plt.imshow(stretch_hist(image1), cmap='gray', vmin=0, vmax=255)
# plt.figure()
# plt.plot(calc_hist(stretch_hist(image1)))
# plt.title('stretched histogram')
```
