```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def equation2(x):
    return 5 - x

def equation3(x):
    return -2 * x

# Generate x values
x = np.linspace(-15, 15, 150)

# Generate y values for each equation
# y1 = equation1(x)
y2 = equation2(x)
y3 = equation3(x)

# Plot the equations
# plt.plot(x, y1, label='y = -5')
plt.plot(x, y2, label='y = 5 - x')
plt.plot(x, y3, label='y = -2x')

# Plot the vertical line x = -5
x_value = -5
plt.axvline(x_value, color='red', linestyle='-', label='x = -5')

# Add labels and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Graphs of Three Equations')
plt.legend()

# Show the plot
plt.grid(True)
plt.show()
```
