Please obtain and display the following modes for color image concatenation.

**Before making changes on the image, it is too big for show when we are joinning it vertically or horizatally, so we Resize it

```python
smaller_percent = 0.6

# Load the image in opencv
org_img = cv2.imread('ComputerDepartment.jpg')

my_img_shape = org_img.shape
print(f'Shape of I : {my_img_shape}')

# RESIZINg new width and height, based on the "smaller_percent"
desired_width = int(my_img_shape[1]*smaller_percent)
desired_height = int(my_img_shape[0]*smaller_percent)

print("previous width->", my_img_shape[0], ", desired width->", desired_width)
print("previous height->", my_img_shape[1], ", desired height->", desired_height)

# Resize the image
my_img = cv2.resize(org_img, (desired_width, desired_height))

cv2.imshow('Vertical Concatenated Image and Rsized', my_img)
cv2.waitKey(2000)
cv2.destroyAllWindows()

plt.imshow(cv2.cvtColor(my_img, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Turn off axis
plt.show()
```

Mode 1 : The main image should be joined vertically with its vertical invert.

```python
# Inverting the image vertically
vertical_inverted = cv2.flip(my_img, 0)  # 0 for vertical flip

# Concatenate the original image and its vertical invert
result = cv2.vconcat([my_img, vertical_inverted])

# Display the result
cv2.imshow('Vertical Concatenated Image and Rsized', result)
cv2.waitKey(2000)
cv2.destroyAllWindows()

plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Turn off axis
plt.show()
```

Mode 2 : The vertical invert of the original image should be joined vertically with the original image.

```python
# Concatenate the original image and itself
result = cv2.vconcat([my_img, my_img])

# Display the result
cv2.imshow('Vertical Concatenated Image and Rsized', result)
cv2.waitKey(2000)
cv2.destroyAllWindows()

plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Turn off axis
plt.show()
```

Mode 3 : The main image should be joined horizontally with its horizontal invert.

```python
# Invert the image horizontally
horizontal_inverted = cv2.flip(my_img, 1)  # 1 for horizontal flip

# Concatenate the original image and its horizontal invert
result = cv2.hconcat([my_img, horizontal_inverted])

# Display the result
cv2.imshow('Horizontal Concatenated Image', result)
cv2.waitKey(2000)
cv2.destroyAllWindows()

plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Turn off axis
plt.show()
```

Mode 4 : The horizontal invert of the original image should be joined horizontally with the original image.

```python
# Concatenate the original image and itself
result = cv2.hconcat([my_img, my_img])

# Display the result
cv2.imshow('Horizontal Concatenated Image', result)
cv2.waitKey(2000)
cv2.destroyAllWindows()

plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.axis('off')  # Turn off axis
plt.show()
```

First, read and display the image of the iust logo. Then show each of its color channels so that the red channel is displayed in the blue, green channel in the green, and red channel in red. Notice that since these are color images they have to be mxnx3 arrays. For a purely red image, the green and blue channels are entirely zero.

Read logo and show it.

```python
# Load the original image
logo_image = cv2.imread('logo.jpg')

# Display the result
cv2.imshow('Windows Colored Image', logo_image)
cv2.waitKey(2000)
cv2.destroyAllWindows()
```

Show red channel.

```python
# Split the image into its color channels
b, g, r = cv2.split(logo_image)

# Create a new image with only the blue channel
blue_channel_image = cv2.merge([b, np.zeros_like(g), np.zeros_like(r)])

# Display the original and blue channel images using matplotlib
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(logo_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Blue Channel')
plt.imshow(cv2.cvtColor(blue_channel_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()

# Display the modified image
cv2.imshow('Blue Channel Logo Image', blue_channel_image)
cv2.waitKey(2000)
cv2.destroyAllWindows()

# Save the resulted image
cv2.imwrite('Logo_blue_channel.jpg', blue_channel_image)
```

Show green channel.

```python
# Split the image into its color channels
b, g, r = cv2.split(logo_image)

# Create a new image with only the green channel
green_channel_image = cv2.merge([np.zeros_like(b), g, np.zeros_like(r)])

# Display the original and green channel images using matplotlib
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(logo_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Green Channel')
plt.imshow(cv2.cvtColor(green_channel_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()

# Display the modified image
cv2.imshow('Green Channel Logo Image', green_channel_image)
cv2.waitKey(2000)
cv2.destroyAllWindows()

# Save the resulted image
cv2.imwrite('Logo_green_channel.jpg', green_channel_image)
```

Show red channel.

```python
# Split the image into its color channels
b, g, r = cv2.split(logo_image)

# Create a new image with only the red channel
red_channel_image = cv2.merge([np.zeros_like(b), np.zeros_like(g), r])

# Display the original and red channel images using matplotlib
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(logo_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title('Red Channel')
plt.imshow(cv2.cvtColor(red_channel_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

plt.show()

# Display the modified image
cv2.imshow('Red Channel Logo Image', red_channel_image)
cv2.waitKey(2000)
cv2.destroyAllWindows()

# Save the resulted image
cv2.imwrite('Logo_red_channel.jpg', red_channel_image)
```
